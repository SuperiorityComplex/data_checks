```python
# Initialize database connection
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from src.data_checks.database.utils.engine import connect
from src.data_checks.database.managers.utils import session_utils
from src.data_checks.database.managers.models import *
from src.data_checks.database.managers import *

check_database_url = "postgresql://ivanzhang:@localhost:5432/test_data_checks"

if check_database_url:
    engine = connect(check_database_url)
    Base.metadata.create_all(engine)
    session_utils.configure(engine)
```

```python
# RuleExecutions
from datetime import datetime, timedelta

with session_utils.session_scope() as session:
    # Get all executions from the last day
    executions = session.query(RuleExecution).filter(
        RuleExecution.created_at > datetime.now() - timedelta(days=1)
    ).all()
    print(executions)

    for execution in executions:
        print(execution.rule.name, execution.params)
        if (execution.status == "failure"):
            print("FAILED: ", execution.exception)
```

```python
# Rules
with session_utils.session_scope() as session:
    # Get all the checks and their code
    rules = session.query(Rule).all()
    print(rules)

    for rule in rules:
        print(rule.name)
        print(rule.executions)
        print(rule.code)
```
